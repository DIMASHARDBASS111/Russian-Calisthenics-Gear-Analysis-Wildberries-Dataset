```python
import pandas as pd, ast
```

```python
df = pd.read_csv('WorkoutProducts.csv')

df['category'].value_counts()
```

```python
df.columns
```

```python
df = df[['id', 'name', 'brand', 'category', 'rating', 'feedbacks', 'supplierRating', 'sizes']]
df.iloc[0]['sizes']
```

```python
df.info()
```

```python
df['sizes'] = df['sizes'].apply(lambda x: ast.literal_eval(x))

type(df.iloc[0]['sizes'])
```

```python
df['base_price'] = df['sizes'].apply(lambda x: x[0]['price']['basic']/100.0)
df['discount_price'] = df['sizes'].apply(lambda x: x[0]['price']['product']/100.0)
```

```python
df.head()
```

```python
df = df.drop(columns='sizes')
```

```python
df.to_csv('Cleaned_Workout_Products.csv')
```
