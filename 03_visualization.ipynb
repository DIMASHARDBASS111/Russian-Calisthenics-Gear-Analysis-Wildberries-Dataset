```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns

df = pd.read_csv('Cleaned_Workout_Products.csv')
df
```

```python
df['discount'] = round((df['base_price'] - df['discount_price'])*100/df['base_price'], 2)

df['revenue_estimated'] = df['discount_price'] * df['feedbacks']

df
```

```python
categories = ['турник', 'пояс с цепью для отягощения', 'упоры для отжиманий', 'фитнес резинки', 'брусья', 'гимнастические кольца']
sns.set_theme(style="darkgrid", palette="deep")
fig, ax = plt.subplots(2,3, figsize=(12, 6))

for i in range(0, len(categories)):
    sns.boxplot(data=df[df['category'] == categories[i]], x = 'discount_price', ax=ax[i%2][i%3], orient = 'h', width=.2)
    ax[i%2][i%3].set_xlabel('')
    ax[i%2][i%3].set_ylabel(categories[i])
    ax[i%2][i%3].set_xlim(0, 8000)
    ax[i%2][i%3].grid(axis="x", linestyle="--", alpha=0.3, color="black")
fig.tight_layout()
```
